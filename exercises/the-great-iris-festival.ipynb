**Exercise**

Explore the **Iris dataset** and learn about the distributions and relationships of its features using Seaborn to help the judges decide which species is the most talented of them all!

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris_df = sns.load_dataset('iris')

iris_df.head()
```

```python
sns.pairplot(iris_df, hue="species")
```

**Task 1: Distribution of Sepal Length**

1. Create a histogram of the `sepal_length` column using `sns.histplot`.
2. Color the histogram using the `species` column.
3. Customize the appearance (color, bin size, etc.).

```python
sns.histplot(iris_df, x = 'sepal_length', hue='species', bins=20, kde=True)
plt.show()
```

**Task 2: Compare Sepal Width and Length**
1. Use `sns.scatterplot` to compare `sepal_width` and `sepal_length`.
2. Color the points based on the species.
3. Add a title and labels for clarity.


```python
sns.scatterplot(iris_df, x = 'sepal_width', y = 'sepal_length', hue = 'species')
```

**Task 3: Pairwise Relationships**
1. Use `sns.pairplot` to visualize pairwise relationships between all numerical columns.
2. Color the points based on species.
3. Do you observe any clusters? Which features separate the species the best?

```python
sns.pairplot(iris_df, hue='species')
```

**Task 4: Boxplots for Comparing Features**

1. Use `sns.boxplot` to showcase the distribution of `petal_length` across different species.
2. Observe which species has the widest range of `petal_length` and which one has the smallest.

```python
sns.boxplot(iris_df, x = 'species', y = 'petal_length')

plt.show
```

**Task 5: Violin Plot Exploration**

1. Create a `sns.violinplot` comparing the distribution of `petal_width` across different species.
2. Contrast this with the box plot. What additional information can you derive from the violin plot?

```python
sns.violinplot(iris_df, x = 'species', y = 'petal_length')

plt.show
```
