```python
import matplotlib.pyplot as plt

import seaborn as sns

import numpy as np

ax_0 = plt.subplot(2, 1, 1)  # 2 rows, 1 column, index 1

ax_1 = plt.subplot(2, 1, 2)  # Si yo creo otra subplot se genera con 2 filas y 1 columna. 

# En una figura es posible generar distintos subplots

x = np.arange(0,10,0.1)

y = x**2

x,y  # Para que se pueda pintar tiene que haber el mismo numero de elementos en x e y.

z = np.cos(x)

ax_0.plot(x,y) # Esto me pinta la funcion cuadratica

# ax_0.scatter(x,y) # Esto me pinta los puntos y la combinacion de los dos me devuelve un grafico de funcion cuadratica con puntos

ax_1.plot(x,z)

# Las funciones siempre devuelven algo, aunque sea un objeto vacio

# En este caso esta grafica es la representacion de la relacion entre un numero y su cuadratico
```

```python
ax = plt.subplot()

x = np.arange(-5,5,0.1)

y_0 = x**2

y_1 = np.cos(x)

ax.plot(x, y_0, "r--")

ax.plot(x, y_1, "b-+")

ax.legend(["y = x^2", "y = cos(x)"])

ax.set_title("This represents y=x^2")

ax.set_xlabel("x(s)")

ax.set_ylabel("y(m)")



```

```python
ax = plt.subplot()

x = np.arange(-5,5,0.5)

y_0 = x**2

y_1 = np.cos(x)

ax.scatter(x, y_0)

ax.scatter(x, y_1)

ax.legend(["y = x^2", "y = cos(x)"])

ax.set_title("This represents y=x^2")

ax.set_xlabel("x(s)")

ax.set_ylabel("y(m)")
```

```python
ax = plt.subplot()

x = np.arange(0,5,1)

y_0 = x**2

y_1 = np.cos(x)

ax.bar(x, y_0)

ax.bar(x, y_1)

ax.legend(["y = x^2", "y = cos(x)"])

ax.set_title("This represents y=x^2")

ax.set_xlabel("x(s)")

ax.set_ylabel("y(m)")
```
